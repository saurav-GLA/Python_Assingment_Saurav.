.# PART 4: Students Performance

```python
# 1. Display average marks in math, reading, and writing.
avg_math = students["math score"].mean()
avg_reading = students["reading score"].mean()
avg_writing = students["writing score"].mean()

print("Average Math Score:", avg_math)
print("Average Reading Score:", avg_reading)
print("Average Writing Score:", avg_writing)
```

```python
# 2. Who performed better in writing – males or females?
writing_by_gender = students.groupby("gender")["writing score"].mean()
print("Average Writing Score by Gender:\n", writing_by_gender)

better_gender = writing_by_gender.idxmax()
print("\nBetter performer in writing:", better_gender)
```

```python
# 3. Add a new column for total score and percentage.
students["total_score"] = students["math score"] + students["reading score"] + students["writing score"]
students["percentage"] = students["total_score"] / 3 
students.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# 4. Plot a bar chart showing average scores grouped by parental level of education.
avg_scores_by_parent = students.groupby("parental level of education")[["math score", "reading score", "writing score"]].mean()

avg_scores_by_parent.plot(kind='bar', figsize=(10,6))
plt.title("Average Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xlabel("Parental Level of Education")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Find students who scored above 90% in al subjects.
top_students = students[
    (students["math score"] > 90) &
    (students["reading score"] > 90) &
    (students["writing score"] > 90)
]

print("Students with >90% in all subjects:\n", top_students)
```
